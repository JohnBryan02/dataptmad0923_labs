For this lab, we are going to be putting together a user similarity based recommender system in a step-by-step fashion. Our data set contains customer grocery purchases, and we will use similar purchase behavior to inform our recommender system. Our recommender system will generate 5 recommendations for each customer based on the purchases they have made.

```python
import pandas as pd
from scipy.spatial.distance import pdist, squareform
```

```python
data = pd.read_csv('../data/customer_product_sales.csv')
```

```python
data.head()
```

## Step 1: Create a data frame that contains the total quantity of each product purchased by each customer.

You will need to group by CustomerID and ProductName and then sum the Quantity field.

```python
df = pd.DataFrame(data)
```

```python
df = df.groupby(['CustomerID','ProductName'])['Quantity'].sum().reset_index()
df
```

## Step 2: Use the `pivot_table` method to create a product by customer matrix.

The rows of the matrix should represent the products, the columns should represent the customers, and the values should be the quantities of each product purchased by each customer. You will also need to replace nulls with zeros, which you can do using the `fillna` method.

```python
table = pd.pivot_table(df,values = 'Quantity', index=['ProductName'],columns=['CustomerID'],fill_value=0)
#table.fillna(0)
table
```

## Step 3: Create a customer similarity matrix using `squareform` and `pdist`. For the distance metric, choose "euclidean."

```python
#primero pdist y luego square form
#plist es un dataframe y luego se transforma a square form
#import scipy df = plist()->squareform()
#metric euclidean
```

```python
#Transponemos la tabla para tener customer vs products
customer_product_matrix = table.transpose()
```

```python
distance = pdist(customer_product_matrix,metric='euclidean')
distance
```

```python
matrix = pd.DataFrame(squareform(distance), index=customer_product_matrix.index, columns=customer_product_matrix.index)
matrix
```

## Step 4: Check your results by generating a list of the top 5 most similar customers for a specific CustomerID.

```python
"""filtro para df(customer_id)--> ordenar(sort)-->Select top5(df.head or iloc)"""
filtered_customer_id = 33
```

```python
similars = matrix.loc[filtered_customer_id]
similars
```

```python
#sorting
top5 = similars.sort_values(ascending=True).head(6)[1:] #Con esto quitamos el primero que es el mismo customerID
top5
```

## Step 5: From the data frame you created in Step 1, select the records for the list of similar CustomerIDs you obtained in Step 4.

```python
#df del step 1 +(merge,join)+ df del step 4 or customerid
#final df (customerid,customername,....)
```

```python
selecting = df.loc[top5.index]
selecting
```

## Step 6: Aggregate those customer purchase records by ProductName, sum the Quantity field, and then rank them in descending order by quantity.

This will give you the total number of each product purchased by the 5 most similar customers to the customer you selected in order from most purchased to least.

```python
#groupby(p.name)sumquantity()
#rank = order or sort by descending of quantity
```

```python
filter_products = selecting.groupby(['ProductName'])['Quantity'].sum().reset_index().sort_values('Quantity',ascending=False)
filter_products
```

## Step 7: Filter the list for products that the chosen customer has not yet purchased and then recommend the top 5 products with the highest quantities that are left.

- Merge the ranked products data frame with the customer product matrix on the ProductName field.
- Filter for records where the chosen customer has not purchased the product.
- Show the top 5 results.

```python
#filter by 0 y recomendar los 5 productos con mayores ventas left??
```

```python
#merge
merged = pd.merge(customer_product_matrix.T.reset_index(),filter_products,how='inner', on='ProductName')
merged
```

```python
recomendation = merged.loc[merged[33] == 0, ['ProductName', 33,'Quantity']].sort_values('Quantity', ascending=False).head()
recomendation
```

## Step 8: Now that we have generated product recommendations for a single user, put the pieces together and iterate over a list of all CustomerIDs.

- Create an empty dictionary that will hold the recommendations for all customers.
- Create a list of unique CustomerIDs to iterate over.
- Iterate over the customer list performing steps 4 through 7 for each and appending the results of each iteration to the dictionary you created.

```python
#crear lista que contenga diccionarios
#un for como una casa
```

```python
#Creamos un diccionario vacio
recomendaciones = {}
#Creamos un listado de los customer id para iterar sobre ellos.
customer_list = table.columns
```

```python
#creamos el loop
for customer_id in customer_list:
    similars = matrix.loc[customer_id]
    top5 = similars.sort_values(ascending=True).head(6)[1:]
    selecting = df.loc[top5.index]
    filter_products = selecting.groupby(['ProductName'])['Quantity'].sum().reset_index().sort_values('Quantity', ascending=False)
    merged = pd.merge(customer_product_matrix.T.reset_index(), filter_products, how='inner', on='ProductName')
    recommendation = merged.loc[merged[customer_id] == 0, ['ProductName', 'Quantity']].sort_values('Quantity', ascending=False).head()
    recomendaciones[customer_id] = recommendation
```

##  Step 9: Store the results in a Pandas data frame. The data frame should a column for Customer ID and then a column for each of the 5 product recommendations for each customer.

```python
#Creamos el dataframe en blanco
recomendaciones_df = pd.DataFrame()
#aÃ±adimos los ids
recomendaciones_df['CustomerID'] = customer_list
recomendaciones_df
```

```python
#Rellenamos el diccionario
#ESTO ESTA GENERADO CON CHATGPT por que si no me vuelvo loco.
for customer_id, recommendation in recomendaciones.items():
    # Crea columnas dinamicas para cada recomendaciÃ³n.
    cols = [f'Recommendation_{i+1}' for i in range(len(recommendation))]
    
    #Rellenamos las nuevas columnas.
    recomendaciones_df.loc[recomendaciones_df['CustomerID'] == customer_id, cols] = recommendation['ProductName'].values

# Cambiamos el tipo de nuestro customer id a object. 
recomendaciones_df = recomendaciones_df.astype({'CustomerID': 'object'})

recomendaciones_df
```
